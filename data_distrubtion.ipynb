```python
import plotly.express as px
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("event_mag_phase_with_waveforms.csv") #get waveform data into a dataframe
```

```python
#now going to plot distrubition of magnitude in our data
fig=px.histogram(df['magnitude'], x='magnitude')
fig.update_xaxes(range=[0,5])

fig.show()
```

```python
std = np.std(df['magnitude'])
mean = np.mean(df['magnitude'])
print("standard deviation is", std,
      "the mean is", mean)
```
